# Config

### Numpy

By default, ```numpy``` prints floating-point numbers in a shortened format. <br /> 
When they are simple enough, omitting trailing zeros makes no difference <br /> 
to the values precision. We can change this using the numpy settings to <br /> 
update the print options.

````python
plt.style.use('dark_background')

# Print with full precision and trailing zeros
np.set_printoptions(precision=1, floatmode='fixed')
````

````python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
````

````python
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
````

# Acquiring "M"

```m``` is representative of the ```total``` number of examples we have in a training set. <br />
To acquire ```m``` we can use two methods, we can either attain the shape of a ```numpy``` <br />
array, or we can acquire the ```len```.

````python
print(f"x_train.shape = {x_train.shape}")
m = x_train.shape[0]
print(f"No. of training examples is: {m}")
````

````python
m = len(x_train)
print(f"No. of training examples is: {m}")
````

# Training Example

We use (x<sup>(i)</sup>,y<sup>(i)</sup>) to denote the <bold>i<sup>th</sup></bold> training example.

## Zip

Combines x_train and y_train into pairs of values. This creates an iterator of tuples where each tuple contains one element from x_train and one from y_train.

## Enumerate

Adds a counter (index) to the tuples produced by zip(), so you get both the index and the tuple of values.


````python
for i, (x_i, y_i) in enumerate(zip(x_train, y_train)):
    print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
````

````python
# Plot the points
plt.scatter(x_train, y_train, marker='x', color='r')

# Set the title 
plt.title('Housing Prices')

# Set the x-axis label
plt.xlabel('Size (1000 sqft)')

# Set the y-axis label
plt.ylabel('Price (1000 USD)')

plt.show()
````

````python
# This produces a line which does not fit the data well
w = 100
b = 100
````

````python
def compute_model_output(x, w, b):
    """
    Computes the predictions of a linear model
    Args:
      x (ndarry, (m,)): Data, m examples
      w,b (scalar)    : Model parameters
    Returns:
     f_wb (ndarray (m,)): Model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m): 
      f_wb[i] = w*x[i] + b
    return f_wb
````

````python
w = 200 # changes the slope of the line
b = 100 # shifts line up and down on y-axis, changes the intersect

tmp_f_wb = compute_model_output(x_train, w, b)

# Plot the model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', color='r', label='Actual Values')

# Set the title 
plt.title('Housing Prices')

# Set the x-axis label
plt.xlabel('Size (1000 sqft)')

# Set the y-axis label
plt.ylabel('Price (1000 USD)')

plt.legend()
plt.show()
````

````python
# Now that we haev a model we can use it to make an original prediction. 

x_i = 1.2
cost_1200sqft = w * x_i + b 

print(f"${cost_1200sqft:.0f} thousand dollars")
````
